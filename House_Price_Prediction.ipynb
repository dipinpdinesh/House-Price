Lets convert the string values to integer by Labelencoder,here we need to convert the below attributes to numbers: **street, city, statezip**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
label.fit(data.street)
```

```python
data.street=label.transform(data.street)
```

```python
data.city=label.fit_transform(data.city)
data.statezip=label.fit_transform(data.statezip)
```

**Lets apply Linear regression method to find the relation between the inputs factors and /Price**


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('price', axis=1)
```

```python
y=data.price
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10) #dividing the entire dataset to testing and training parts
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
regr=LinearRegression()
```

```python
regr.fit(x_train,y_train) # Training the model
```

```python
result = regr.predict(x_test) # Predicting the price value for testing output
```

```python
regr.score(x_test,result)
```

```python
regr.coef_
```

```python
regr.intercept_
```

**Ploting the graph showing the relationship between price and different factors**

```python
plt.figure(figsize=(10,10))
plt.plot(x_test,regr.coef_[0]*x_test+regr.intercept_, color='blue')

plt.xlabel('Factors')
plt.ylabel('Price')
plt.title('House Price Prediction', color='red')
plt.show()

```
